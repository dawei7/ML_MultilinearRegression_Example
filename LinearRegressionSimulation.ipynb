```python
#Loss function

def loss_func(X,y,theta):

    m = len(y) #number of rows
    predictions = np.dot(X,theta)

    return (1/(2*m)) * np.sum((predictions - y)**2)

my_loss = loss_func(X,y,theta)
my_loss
```
